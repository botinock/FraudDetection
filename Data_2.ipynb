```python
df_g['flag'] = df_g['flag_rare'] = 0
df_g.loc[df_g['level_reached_5'] > df_g['level_reached_3'], 'flag'] = 1
df_g.loc[df_g['install'] == 0, 'flag'] = 1
df_g.loc[df_g['NPU'] > 2 | df_g['NPU'] == 1, 'flag_rare'] = 1
df_g.loc[df_g['level_reached_3'] > 2, 'flag_rare'] = 1
df_g.loc[df_g['level_reached_5'] > 2, 'flag_rare'] = 1
df_g
```

```python
for name, tdf in df_g.groupby(level=0):
    print(name)
    print(len(tdf[tdf['level_reached_3'] == 0].index) / len(tdf.index))
```

```python
for name, tdf in df_g.groupby(level=1):
    print(name)
    print(len(tdf[tdf['level_reached_3'] == 0].index) / len(tdf.index))
```

```python
i = 0
for name, tdf in df_g.groupby(level=2):
    x = len(tdf[tdf['level_reached_3'] == 0].index) / len(tdf.index)
    if x > 0.9:
        i = i + 1
        print(name)
        print(x)
print(i)
```

```python
for name, tdf in df_g.groupby(level=0):
    print(name)
    print(len(tdf[tdf['NPU'] > 0].index) / len(tdf.index))
```

```python
for name, tdf in df_g.groupby(level=1):
    print(name)
    print(len(tdf[tdf['NPU'] > 0].index) / len(tdf.index))
```

```python
i = 0
for name, tdf in df_g.groupby(level=2):
    x = len(tdf[tdf['NPU'] > 0].index) / len(tdf.index)
    if x < 0.001 or x == 1:
        i = i + 1
        print(name)
        print(x)
print(i)
```

```python
df_g['flag_zero'] = df_g['flag_NPU'] = 0
for i in range(0, 3):
    for name, tdf in df_g.groupby(level=i):
        x = len(tdf[tdf['level_reached_3'] == 0].index) / len(tdf.index)
        if x == 1:
            df_g.loc[tdf.index, 'flag_zero'] = 1
        x = len(tdf[tdf['NPU'] > 0].index) / len(tdf.index)
        if x < 0.001:
            df_g.loc[tdf.index, 'flag_NPU'] = 1
df_g
```

```python
print(df_g['flag'].sum())
print(df_g['flag_zero'].sum())
print(df_g['flag_NPU'].sum())
```
